# Sales analysis

#### Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

### Task 1: merging 12 months of sales data into a single file

```python
directory_path = 'Sales_Data'
dataframes = []

for file_name in os.listdir(directory_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(directory_path, file_name)
        df = pd.read_csv(file_path)
        dataframes.append(df)

df = pd.concat(dataframes, ignore_index=True)

df
```

```python
df.to_csv("Output/Sales_2019.csv", index=False)
```

### Task 2: add a month column

```python
df['Order Date'].dtype
```

```python
df['Order Date'] = df['Order Date'].astype(str)
df['Month'] = df['Order Date'].apply(lambda x: x[0:2])

df
```

```python
df['Month'].unique()
```

##### Find rows in which 'Month' column has 'na' value

```python
na_month_rows = df[df['Month']=='na']
na_month_rows
```

##### Find rows with 'Or' value in the 'Month' column

```python
Or_month_rows = df[df['Month']=='Or']
Or_month_rows
```

##### Drop those rows

```python
df = df.drop(na_month_rows.index)
df = df.drop(Or_month_rows.index)
```

```python
df['Month'].unique()
```

```python
#convert 'Month' column dtype from object to integer
df['Month'] = df['Month'].astype('int8')
```

### Task 3: add a sales column

##### Convert dtype of 'Quantity Ordered' from string to integer

```python
df['Quantity Ordered'].dtype
```

```python
df['Quantity Ordered'].unique()
```

```python
df['Quantity Ordered'] = df['Quantity Ordered'].astype('int64')
df['Quantity Ordered'].dtype
```

##### Convert dtype of 'Price Each' from string to float

```python
df['Price Each'].dtype
```

```python
df['Price Each'] = df['Price Each'].astype('float64')
df['Price Each'].dtype
```

##### Finally creating the sales column 

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
```

```python
df.head()
```

### Task 4: add a city column

```python
#get city name example
df.iloc[0]['Purchase Address'].split(',')[1].strip()
```

```python
#get state code example
df.iloc[0]['Purchase Address'].split(',')[2].split(' ')[1]
```

```python
def get_city(address):
    return address.split(',')[1].strip().title()


def get_state(address):
    return address.split(',')[2].split(' ')[1]


df['City'] = df['Purchase Address'].apply(lambda x: f'{get_city(x)} ({get_state(x)})')

df
```

```python
df['City'].unique()
```

### Task 5: Create Hour and Minute columns

```python
df['Order Date'].dtype
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Order Date'].dtype
```

```python
df.head()
```

```python
df['Hour'] = df['Order Date'].dt.hour
df['Minute'] = df['Order Date'].dt.minute
df.head()
```

### Question 1: what was the best month for sales? How much was earned that month? 

```python
#calculate the total sales for each month
monthly_sales = df.groupby('Month').sum()

#create a bar chart to visualize the monthly sales
sns.set_style('dark')
ax = sns.barplot(data=monthly_sales, x=monthly_sales.index, y='Sales', color='darkblue')
ax.set_xlabel('Month')
ax.set_ylabel('Sales (millions USD)')

#display bar chart
plt.show()
```

```python
monthly_sales
```

**Answer: december (12), 4.613.443,34 USD**

### Question 2: what city had the highest number of sales?

```python
cities = df.groupby('City').sum()
cities
```

```python
#create a bar chart to visualize sales per city
ax = sns.barplot(data=cities, x=cities.index, y='Sales', color='darkblue')
ax.set_xlabel('City')
ax.set_ylabel('Sales (millions USD)')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")

plt.show()
```

**Answer: San Francisco - CA had the highest number of sales in 2019**

### Question 3: what time should we display advertisiments to maximize likelihood of customer's buying product?

```python
hours = df.groupby('Hour').count()
hours
```

```python
plt.plot(hours.index, hours['Quantity Ordered'])
plt.xticks(hours.index)
plt.xlabel('Hours')
plt.ylabel('# Orders')
plt.grid()
plt.show()
```

**Answer: it seems that we have a local maximum at 12h and a global maximum at 19h, so it might be a good idea to display some advertisiments around those hours**

### Question 4: what products are most often sold together?

```python
duplicated_orders = df[df['Order ID'].duplicated(keep=False)]

duplicated_orders
```

```python
duplicated_orders.loc[:, 'Grouped'] = duplicated_orders.groupby('Order ID')['Product'].transform(lambda x: ', '.join(x))

duplicated_orders
```

```python
orders = duplicated_orders[['Order ID', 'Grouped']].drop_duplicates()

orders
```

```python
# referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary 
from itertools import combinations
from collections import Counter

count = Counter()

#get the combinations of two items
for row in orders['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

#print the 10 most common combinations of two items
count.most_common(10)
```

```python
count2 = Counter()

#get the combinations of three items
for row in orders['Grouped']:
    row_list = row.split(',')
    count2.update(Counter(combinations(row_list, 3)))

#print the 10 most common combinations of three items
count2.most_common(10)
```
